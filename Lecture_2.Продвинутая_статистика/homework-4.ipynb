Написать **линейную регрессию** на два атрибута. Допустим, мы хотим предсказать цену дома (MEDV) в Бостоне, зная что цена на дом имеет линейную зависимость с атрибутом LSTAT. 



1.   Разделить данные на 80% тренировочных, 20% тестовых
2.   Составить модель линейной регрессии
3.   Посчитать ошибку модели на тестовых данных
4.   Сделать точечный график данных и полученной прямой линейной регрессии



```python
plt.scatter(data['MEDV'], data['LSTAT'])
```

```python
medv = data['MEDV'].values.reshape(-1,1)
lstat = data['LSTAT'].values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
model = LinearRegression()

X_train, X_test, y_train, y_test = train_test_split(medv, lstat, test_size = 0.2)
```

```python
model.fit(X_train, y_train)
```

```python
predicted_y = model.predict(X_test)
```

```python
print(mean_squared_error(predicted_y, y_test))
print(mean_absolute_error(predicted_y, y_test))
```

```python
plt.xlabel('medv')
plt.ylabel('lstat')
plt.title('Линейная регрессия')
x = np.linspace(0, 50, 100)
y = model.coef_*x + model.intercept_
y = np.transpose(y)
plt.plot(x,y)
plt.plot(X_train, y_train, 'r.')
plt.plot(X_test, y_test, 'go')
plt.xlim(0,50)
plt.ylim(0,50)
plt.legend(['Прямая линейной регрессии', 'Тренировочные данные', 'Тестовые данные'])
plt.show()
```

```python
number_of_sums = 1000
X1 = np.zeros([number_of_sums])
for i in range(number_of_sums):
    X1[i] = np.sum(np.random.rand(100))/100
plt.hist(X1, 100, facecolor = 'green')

#График изменился т.к к нему применяется закон больших чисел. На графике видно, что он имеет распределение,
#близкое к нормальному, поэтмоу к нему можно применить ЦПТ.
```
